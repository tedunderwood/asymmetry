```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import glob, math, random
from scipy.stats import pearsonr, zscore, ttest_ind
from statistics import mean, stdev
from math import sqrt
```

```python
zdata = pd.read_csv('../supp2results/zdata.tsv', sep = '\t', index_col = 'docid')
```

```python
norton = pd.read_csv('../meta/canon/britnorton.tsv', sep = '\t', index_col = 'docid')
```

```python
testset = [x for x in norton.index.tolist() if str(x) != 'nan']
```

```python
ctr = 0
for t in testset:
    if t in zdata.index:
        print(t, ctr +1)
        ctr = ctr + 1
    else:
        print("NO: ", t)
```

```python
def get_contrast_set(primaryset, data):
    ''' Accepts a list of categories and finds volumes matching those categories in 
    *hypmeta*. Then constructs a set of matching volumes.
    '''
    
    dates_of_cat = data.loc[primaryset, 'latestcomp']
    contrast_vols = []
    
    for d in dates_of_cat:
        population = data.index[data['latestcomp'] == d].tolist()
        population = set(population) - set(primaryset)
        if len(population) < 1:
            print('*', d)
            continue
        else:
            chosen = random.sample(population, 1)[0]
            contrast_vols.append(chosen)
    
    return contrast_vols

contrastset = get_contrast_set(testset, zdata)
```

```python
def general_test(acat, bcat, zdata): 
    ''' Given two sets of volumes, calculates t tests and
    Cohen's d for difference of means.
    '''
    
    global columns
    for col in columns:
        a = zdata.loc[acat, col]
        a = a[~np.isnan(a)]
        b = zdata.loc[bcat, col]
        b = b[~np.isnan(b)]
        t, p = ttest_ind(a, b)
        print(col, "t-test", t, p)
        cohens_d = (mean(a) - mean(b)) / (sqrt((stdev(a) ** 2 + stdev(b) ** 2) / 2))
        print(col, "Cohen's d", cohens_d)
        print()
```

```python
columns = [x for x in zdata.columns.tolist() if x.startswith('resonance')]
```

```python
general_test(testset, contrastset, zdata)
```

Even at n == 23, those are significant effect sizes.
