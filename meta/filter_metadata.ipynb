# Filtering the metadata sample

There are some rows in the initial sample that either have the wrong docids, or correspond to very short files that got cut from the sample. Subsequent processing will be simpler if we correct these errors at the source.

```python
import pandas as pd
```

```python
meta = pd.read_csv('topicsample.tsv', sep = '\t')
meta.head()
```

```python
data = pd.read_csv('../model/doc_topics.tsv', sep = '\t')
data.head()
```

```python
import sys
sys.path.append('/Users/tunder/Dropbox/python/asymmetry/')
import SonicScrewdriver as utils
```

```python
sampledocs = set(meta.docid)
havedocs = set(data['index'])
missdocs = sampledocs - havedocs
print(len(missdocs))
```

```python
stillmiss = []
surplus = havedocs - sampledocs
print('surplus: ', len(surplus))
for d in missdocs:
    cleand = utils.clean_pairtree(d)
    if cleand in havedocs:
        surplus.remove(cleand)
        continue
    else:
        stillmiss.append(d)

print(len(stillmiss))
print('surplus: ', len(surplus))  
```

```python
meta = meta.assign(position = meta.index)
meta.set_index('docid', inplace = True)
meta.head()
```

```python
print(meta.shape)
meta2 = meta.drop(labels = stillmiss, axis = 0)
print(meta2.shape)
```

```python
meta2.to_csv('filteredsample.tsv', sep = '\t', index_label = 'docid')
```
