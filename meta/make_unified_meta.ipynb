# make_unified_metadata

```python
import pandas as pd
```

```python
main = pd.read_csv('lastnamesample.tsv', sep = '\t')
main.head()
```

```python
supp = pd.read_csv('../supplement1/lastnamesupp1meta.tsv', sep = '\t')
supp.head()
```

Let's reduce both data frames to their shared columns:

*docid, author, title, lastname, earliestdate,* and *actualdate*

In **main**, *inferreddate* will be taken as evidence of both *earliestdate* and *actualdate.*

```python
newmain = main[['docid', 'author', 'lastname', 'inferreddate', 'title']]
newmain = newmain.rename(index = str, columns = {'inferreddate': 'actualdate'})
newmain = newmain.assign(earliestdate = newmain.actualdate)
newmain.head()
```

```python
newsupp = supp[['docid', 'author', 'lastname', 'actualdate', 'earliestdate', 'title']]
newsupp.head()
```

```python
unified = pd.concat([newmain, newsupp], verify_integrity = True)
print(unified.shape)
unified.head()
```

```python
unified.drop_duplicates(subset = ['docid'], inplace = True)
unified.shape
```

```python
unified = unified[['docid', 'author', 'lastname', 'actualdate', 'earliestdate', 'title']]
```

```python
unified.to_csv('unifiedsample.tsv', sep = '\t', index = False)
```
