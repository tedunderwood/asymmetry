# Clean meta for supplement 1

```python
import pandas as pd
```

```python
meta = pd.read_csv('supplement1meta.tsv', sep = '\t')
meta.head()
```

Only keep certain columns.

```python
new = meta[['docid', 'earliestdate', 'actualdate', 'tags', 'author', 'title', 'gender', 'nationality', 'authsvols']]
```

Create a column for lastname only.

```python
def full2last(name):
    global ctr
    if pd.isnull(name):
        return 'anonymous'
    names = name.split()
    last = names[0].strip(', .[]()-')
    if len(last) > 1:
        return last
    else:
        ctr += 1
        return "anonymous" + '_' + last
    
ctr = 0
new = new.assign(lastname = new.author.map(full2last))
print(ctr)
new.head()
```

Drop rows with duplicate docids. By default this will keep the first occurrence, which is the behavior we want, since we want to keep the metadata from prestigeficmeta.

```python
print(new.shape)
new.drop_duplicates(subset = ['docid'], inplace = True)
new.shape
```

```python
new.to_csv('editedsupplement1meta.tsv', sep = '\t', index = False)
```
